```python
import pandas as pd

class MaxDetectionValues:
    
    def __init__(self, max_value):
        self.max_value = max_value
        self.file_list = ['L_freq = 0.1__U_freq = 1__N_phase = 32_and_N_freq = 2048',
                          'L_freq = 1__U_freq = 10__N_phase = 32_and_N_freq = 2048',
                          'L_freq = 10__U_freq = 100__N_phase = 32_and_N_freq = 2048', 
                          'L_freq = 100__U_freq = 1000__N_phase = 32_and_N_freq = 2048']
        
        self.base_path = '/home/astroguy/Desktop/research/research_emannuele/init_coh_search_algthm/sim_pulse_2_standard/'
    
    def DataFrame(self):
        for k in range(len(self.file_list)):
            file_path = self.base_path + self.file_list[k]
            df = pd.read_csv(file_path, delimiter = '\t')
            
            if k == 0:
                df_whole = df
            else:
                df_whole = pd.concat([df_whole, df], ignore_index=True)
        
        df_max = df_whole[df_whole['Detection Statistic'] >= self.max_value]
        return df_max
    
Detection = MaxDetectionValues(2)
Detection.DataFrame()
```
