```python
import matplotlib.pyplot as plt
```

```python
num_of_calculation = [4686528, 18746112, 74984448, 299937792]
numba_time = [0.4527, 0.0551, 0.2245, 0.9227]
general_python_time = [0.0338, 0.0906, 0.3607, 3.8042]

num_of_calculation_gpu = [4686528, 18746112, 74984448, 299937792, 11199751168]
cuda_gpu_time = [2.6257, 0.0543, 0.2117, 0.8443, 3.2988]
```

```python
fig, ax = plt.subplots(figsize = (15, 12))
ax.scatter(num_of_calculation, general_python_time, linewidth = 7, color = 'red')
ax.semilogx(num_of_calculation, general_python_time, color = 'green', linewidth = 5, label = 'Regular pythonic code')
plt.axvline(x = 299937792, linestyle = '-.')
for ind1, ind2 in zip(num_of_calculation_gpu, cuda_gpu_time):
    if ind1 in [4686528, 18746112, 74984448, 299937792]:
        plt.text(ind1, ind2, '({}, {})'.format(ind1, round(ind2, 3)), size = 15, ha = 'left', va = 'top')
    else:
        plt.text(ind1, ind2, '({}, {})'.format(ind1, round(ind2, 3)), size = 15, ha = 'right', va = 'top')

ax.scatter(num_of_calculation, numba_time, linewidth = 7, color = 'red')
ax.semilogx(num_of_calculation, numba_time, color = 'blue', linewidth = 5, label = 'Numba accelerated code')
ax.scatter(num_of_calculation_gpu, cuda_gpu_time, linewidth = 7, color = 'red')
ax.semilogx(num_of_calculation_gpu, cuda_gpu_time, color = 'magenta', linewidth = 5, label = 'GPU accelerated code')
ax.set_xlabel('Number of calculation', fontsize = 20, fontweight = 'bold' )
ax.set_ylabel('Time taken(in seconds)', fontsize = 20, fontweight = 'bold')
ax.tick_params(axis='both', which='major', labelsize=20)
ax.legend(loc = 'upper left', fontsize = 15)
ax.set_title('Speed(or efficiency) comparison between accelerated and normal code' , fontsize = 20, fontweight = 'bold')
plt.savefig('/home/ak/Desktop/speedcomapre.jpg')
plt.show()
```
