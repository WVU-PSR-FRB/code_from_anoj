```python
import matplotlib.pyplot as plt
```

```python
num_of_calculation = [4686528, 18746112, 74984448, 299937792]
numba_time = [0.7295937538146973, 0.016390085220336914, 0.016546964645385742, 0.016318321228027344]
general_python_time = [0.06199216842651367, 0.12781214714050293, 0.4987800121307373, 1.959019660949707]
```

```python
fig, ax = plt.subplots(figsize = (15, 12))
ax.scatter(num_of_calculation, general_python_time, linewidth = 7, color = 'red')
ax.semilogx(num_of_calculation, general_python_time, color = 'green', linewidth = 5, label = 'Normal pythonic Way')
for ind1, ind2 in zip(num_of_calculation, general_python_time):
    if ind1 in [4686528, 18746112, 74984448]:
        plt.text(ind1, ind2, '({}, {})'.format(ind1, round(ind2, 3)), size = 15, ha = 'left', va = 'top')
    else:
        plt.text(ind1, ind2, '({}, {})'.format(ind1, round(ind2, 3)), size = 15, ha = 'right', va = 'top')

ax.scatter(num_of_calculation, numba_time, linewidth = 7, color = 'red')
ax.semilogx(num_of_calculation, numba_time, color = 'blue', linewidth = 5, label = 'Numba accelerated way')
for i1 in range(len(numba_time)):
    time_value = round(numba_time[i1], 3)
    plt.text(num_of_calculation[i1], time_value , '{}'.format(time_value), size = 15,ha = 'center', va = 'bottom')
ax.set_xlabel('Number of calculation', fontsize = 20, fontweight = 'bold' )
ax.set_ylabel('Time taken(in seconds)', fontsize = 20, fontweight = 'bold')
ax.tick_params(axis='both', which='major', labelsize=20)
ax.legend(loc = 'upper center', fontsize = 20)
ax.set_title('Speed(or efficiency) comparison between accelerated and normal code' , fontsize = 20, fontweight = 'bold')
plt.savefig('/home/ak/Desktop/speedcomapre.jpg')
plt.show()
```
